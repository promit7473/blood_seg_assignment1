# K-mean Thresholding

```python
import cv2
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
import tensorflow as tf
from tensorflow import keras as K
import skimage
from skimage.io import imshow, imread, imsave
from glob import glob
import os
```

```python
img = cv2.imread('./bloodcells.jpg')
imshow(img)
```

```python
img2 = img.reshape((-1,3))
img2=np.float32(img2)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
k=3
attempts=10
ret,label,center=cv2.kmeans(img2,k,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
center=np.uint8(center)
res = center[label.flatten()]
res2 = res.reshape((img.shape))


imshow(res2)
```

```python
# read an input image as gray
# im_gray = cv2.imread(res2,0)
im_gray = cv2.cvtColor(res2, cv2.COLOR_BGR2GRAY)
    
#blur = cv2.GaussianBlur(im_gray,(5,5),0)

# threshold it with your saved threshold
im_bw = cv2.threshold(im_gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

imshow(im_bw)
```
